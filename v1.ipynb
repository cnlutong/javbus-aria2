```python
import json
import urllib.request

import aria2p


# javbus-api， https: // github.com/ovnrain/javbus-api
# 使用此工具来从javbus.com获取信息。此工具需要docker环境运行。
class Jav:
    def __init__(self, ip, port, other):
        self.ip = ip
        self.port = port
        self.other = other

    def get_jav_from_id(self, id):
        surl = self.ip + self.port + self.other + "movies/" + id
        # print(surl)
        url = urllib.request.urlopen(surl)
        try:
            re = json.load(url)
        except:
            pass
        return re

    def get_jav_from_idol(self, idol, page):
        surl = self.ip + self.port + self.other + "movies?page=" + \
            str(page) + "&starId=" + idol + "&magnet=exist"
        # print(surl)
        url = urllib.request.urlopen(surl)
        return json.load(url)

    def get_jav_from_tag(self, tag, page):
        surl = self.ip + self.port + self.other + "movies?page=" + \
            str(page) + "&tagId=" + tag + "&magnet=exist"
        # print(surl)
        url = urllib.request.urlopen(surl)
        return json.load(url)

# 使用番号下载


def download_from_id(jav, id, only):
    print(id)
    data = jav.get_jav_from_id(id)
    # print(data)
    downlink = ""
    if (only != 0):
        if (len(data["stars"]) > only):
            print("Limite !")
            print()
            return
    for i in data["magnets"]:
        # 默认下载不含有字幕的片源
        if (i["hasSubtitle"] == False):
            downlink = i["link"]
            break
    if (downlink == ""):
        return
    print(downlink)
    download = aria2.add_magnet(downlink)
    print()


# 使用女优ID下载
def download_from_idol(jav, idol, page, only):
    j = 1
    while j <= page:
        data = jav.get_jav_from_idol(idol, j)
        for i in data["movies"]:
            # print(i["id"])
            download_from_id(jav, i["id"], only)
        j = j+1

# 使用标签ID下载


def download_from_tag(jav, tag, page, only):
    j = 1
    while j <= page:
        data = jav.get_jav_from_tag(tag, j)
        for i in data["movies"]:
            # print(i["id"])
            download_from_id(jav, i["id"], only)
        j = j+1

# 正文开始


# 初始化Aria2，填写配置信息
aria2 = aria2p.API(
    aria2p.Client(
        host="",
        port=443,
        secret=""
    )
)

# 初始化javbus-api配置信息
ip = "0"
port = ":13000"
other = "/api/v1/"
jav = Jav(ip, port, other)


# download_from_id(jav, "DVDMS-711", 100)
# download_from_idol(jav, "xix", 3, 3)
download_from_tag(jav, "ef", 3, 100)
```
